```python
import pandas as pd
import numpy as np
```

```python
replacer = lambda str: str.lower().str.replace(' ','_')
df.columns= replacer(df.columns.str)
for col in list(df.dtypes[df.dtypes == 'object'].index):
    df[col] = replacer(df[col].str)
df.head().T
```

```python
df.do_not_email = (df.do_not_email == 'yes').astype(int)
df.do_not_call = (df.do_not_call == 'yes').astype(int)
df.search = (df.search == 'yes').astype(int)
df.magazine = (df.magazine == 'yes').astype(int)
df.do_not_email = (df.do_not_email == 'yes').astype(int)
df.newspaper_article = (df.newspaper_article == 'yes').astype(int)
df.x_education_forums = (df.x_education_forums == 'yes').astype(int)
df.newspaper = (df.newspaper == 'yes').astype(int)
df.digital_advertisement = (df.digital_advertisement == 'yes').astype(int)
df.through_recommendations = (df.through_recommendations == 'yes').astype(int)
df.receive_more_updates_about_our_courses = (df.receive_more_updates_about_our_courses == 'yes').astype(int)
df.update_me_on_supply_chain_content = (df.update_me_on_supply_chain_content == 'yes').astype(int)
df.get_updates_on_dm_content = (df.get_updates_on_dm_content == 'yes').astype(int)
df.i_agree_to_pay_the_amount_through_cheque = (df.i_agree_to_pay_the_amount_through_cheque == 'yes').astype(int)
df.a_free_copy_of_mastering_the_interview = (df.a_free_copy_of_mastering_the_interview == 'yes').astype(int)

df.a_free_copy_of_mastering_the_interview.head()
```

```python
df.head().T
```

```python
replacerSelect = lambda str: str.replace('select',np.isnan)
df.columns= replacerSelect(df.columns.str)
for col in list(df.dtypes[df.dtypes == 'object'].index):
    df[col] = replacerSelect(df[col].str)
df.head().T
```

```python
df = df.fillna('unknown')
df.head().T
```

```python
from sklearn.model_selection import train_test_split
df_train_full, df_test = train_test_split(df, test_size=0.2, random_state=1)

df_train, df_val = train_test_split(df_train_full, test_size=0.33, random_state=1)
y_train = df_train.converted.values
y_val = df_val.converted.values

del df_train['converted']
del df_val['converted']

df_train.head().T
```

```python
global_mean = df_train_full.converted.mean()
round(global_mean, 3)
```

```python
categorical = ['prospect_id', 'lead_origin', 'lead_source', 'totalvisits', 'page_views_per_visit', 
               'last_activity', 'country', 'specialization', 'how_did_you_hear_about_x_education', 
               'what_is_your_current_occupation', 'what_matters_most_to_you_in_choosing_a_course',  
               'tags', 'lead_quality', 'lead_profile', 'city', 'asymmetrique_activity_index', 
               'asymmetrique_profile_index', 'asymmetrique_activity_score', 
               'asymmetrique_profile_score', 'last_notable_activity']

numerical = ['lead_number', 'do_not_email', 'do_not_call', 'total_time_spent_on_website', 'search', 'magazine', 
             'newspaper_article', 'x_education_forums', 'newspaper', 'digital_advertisement', 'through_recommendations', 
             'receive_more_updates_about_our_courses', 'update_me_on_supply_chain_content', 
             'get_updates_on_dm_content','i_agree_to_pay_the_amount_through_cheque', 'a_free_copy_of_mastering_the_interview']

df[numerical].nunique()
```

```python
from sklearn.metrics import mutual_info_score
```

```python
train_dict = df_train[categorical + numerical].to_dict(orient='records')
dict(sorted(train_dict[0].items()))
```

```python
from sklearn.feature_extraction import DictVectorizer

dv = DictVectorizer(sparse=False)
dv.fit(train_dict)
```

```python
X_train = dv.transform(train_dict)
X_train[0]
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(solver='liblinear')
model.fit(X_train, y_train)
```

```python
val_dict = df_val[categorical + numerical].to_dict(orient='records')
X_val = dv.transform(val_dict)
y_pred = model.predict_proba(X_val)
y_pred
```

```python
y_pred = model.predict_proba(X_val)[:, 1]
y_pred
```

```python
converted = y_pred >= 0.5
converted
```

```python
round((y_val == converted).mean(), 3)
```
